```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("/content/netflix_titles.csv")
```

```python
print(np.shape(df))
```

```python
print(np.max(df['release_year']))
```

```python
print(np.min(df['release_year']))
```

```python
sns.set(style="whitegrid", palette="deep")
```

```python
df.fillna("Unknown", inplace=True)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="type", palette=["#0074D9", "#FF4136"])
plt.title("Movies vs TV Shows", fontsize=14)
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df["release_year"] = pd.to_numeric(df["release_year"], errors="coerce")
yearly = df["release_year"].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=yearly.index, y=yearly.values, color="#2ECC40")
plt.title("Titles Released per Year", fontsize=14)
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
countries = df["country"].str.split(",").explode().str.strip()
top_countries = countries.value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="flare")
plt.title("Top 10 Countries by Number of Titles", fontsize=14)
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
genres = df["listed_in"].str.split(",").explode().str.strip()
top_genres = genres.value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=top_genres.values, y=top_genres.index, palette="magma")
plt.title("Top 10 Genres on Netflix", fontsize=14)
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```
